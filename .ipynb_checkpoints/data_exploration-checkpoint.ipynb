```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import os
import warnings

# Ignore routine warnings for cleaner output
warnings.filterwarnings('ignore')

# 1. Load your dataset
try:
    df = pd.read_csv("/explore/nobackup/people/spotter5/anna_v/v2/v2_model_training_alt_soil_lc_co2.csv")
except FileNotFoundError:
    print("Error: The data file was not found. Please check the path.")
    exit()

# 2. Create tmean_C and date columns
df['tmean_C'] = df[['tmmn', 'tmmx']].mean(axis=1)
df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))

# 3. Define predictors and target
predictor_vars = [
    'EVI', 'NDVI', 'sur_refl_b01', 'sur_refl_b02', 'sur_refl_b03',
    'sur_refl_b07', 'NDWI', 'pdsi', 'srad', 'tmean_C', 'vap', 'vs', 'co2_cont']
target_var = 'nee'

# --- 🛰️ Time Series Panel Plotting Section ---
print("--- 📊 Starting Time Series Panel Plot Generation ---")

# 4. Define output path for plots
out_path = os.path.join("/explore/nobackup/people/spotter5/anna_v/v2/exploration", target_var)
panel_plot_path = os.path.join(out_path, "panel_plots_unshared_x") # New folder for these plots
os.makedirs(panel_plot_path, exist_ok=True)

# 5. Find the top 15 sites with the longest time series
site_counts = df['site_reference'].value_counts()
top_15_sites = site_counts.nlargest(15).index.tolist()
print(f"Identified top 15 sites for plotting: {top_15_sites}")

# 6. Loop through each PREDICTOR to create one figure
for predictor in predictor_vars:
    print(f"  -> Creating panel plot for: {predictor} (Unshared X-Axis)")
    
    # Create a figure with a 5x3 grid of subplots. Changed sharex to False.
    fig, axes = plt.subplots(5, 3, figsize=(30, 28), sharex=False)
    axes = axes.flatten()

    # Loop through each of the 15 SITES to populate the subplots
    for i, site in enumerate(top_15_sites):
        ax1 = axes[i]
        site_df = df[df['site_reference'] == site].sort_values('date')

        if site_df.empty or site_df[predictor].isnull().all():
            ax1.text(0.5, 0.5, 'No Data', ha='center', va='center')
            ax1.set_title(f"Site: {site}", fontweight='bold')
            continue
        
        # --- Plotting on the subplot ---
        ax1.plot(site_df['date'], site_df[target_var], color='black', label=target_var, linewidth=2)
        ax1.set_ylabel(target_var, color='black')
        ax1.tick_params(axis='y', labelcolor='black')
        ax1.axhline(0, color='gray', linestyle=':', linewidth=1.5)
        
        ax2 = ax1.twinx()
        ax2.plot(site_df['date'], site_df[predictor], color='tab:red', linestyle='--', label=predictor)
        ax2.set_ylabel(predictor, color='tab:red')
        ax2.tick_params(axis='y', labelcolor='tab:red')
        
        # --- Formatting the subplot ---
        ax1.set_title(f"Site: {site}", fontweight='bold')
        ax1.grid(which='major', linestyle='--', linewidth='0.5', color='grey')
        
        # Add X-axis label to every plot now that axes are independent
        ax1.set_xlabel("Date")

    # --- Formatting the entire FIGURE ---
    fig.suptitle(f"Time Series of {target_var} vs. {predictor} Across Top 15 Sites", fontsize=24, y=0.99)

    legend_elements = [
        Line2D([0], [0], color='black', lw=2, label=target_var),
        Line2D([0], [0], color='tab:red', linestyle='--', lw=2, label=predictor)
    ]
    fig.legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(0.95, 0.97), fontsize=14)
    
    fig.tight_layout(rect=[0, 0, 1, 0.96])

    plot_filename = f"{predictor}_panel_plot_unshared_x.png"
    plt.savefig(os.path.join(panel_plot_path, plot_filename), dpi=150)
    plt.close(fig)

print(f"\n--- ✅ Finished plotting. Plots are saved in: {panel_plot_path} ---")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import warnings

# Ignore routine warnings for cleaner output
warnings.filterwarnings('ignore')

# 1. Load your datasets
print("--- 📂 Loading datasets ---")
try:
    df_new = pd.read_csv("/explore/nobackup/people/spotter5/anna_v/v2/v2_model_training_alt_soil_lc_co2.csv")
    df_new = df_new[df_new['flux_method'] == 'EC']
    df_new['tmean_C'] = df_new[['tmmn', 'tmmx']].mean(axis=1)

    df_old = pd.read_csv("/explore/nobackup/people/spotter5/anna_v/v2/v2_model_training_data_alt_soil_old.csv")
    df_old['tmean_C'] = df_old[['tmax_C', 'tmin_C']].mean(axis=1)

    # df_old = df_old[df_old['flux_method'] == 'EC']

except FileNotFoundError as e:
    print(f"Error: A data file was not found. Please check the path.\n{e}")
    exit()

# 2. Prepare dataframes for joining, including NEE from both
new_subset = df_new[['site_reference', 'year', 'month', 'nee', 'NDVI', 'EVI']].rename(
    columns={'NDVI': 'NDVI_new', 'EVI': 'EVI_new'}
)
# --- IMPORTANT: Include 'nee' from the old file as well ---
old_subset = df_old[['site_reference', 'year', 'month', 'nee', 'NDVI', 'EVI']].rename(
    columns={'NDVI': 'NDVI_old', 'EVI': 'EVI_old'}
)

# 3. Join the dataframes. This will create 'nee_x' (from new) and 'nee_y' (from old)
print("--- 🔗 Merging old and new dataframes ---")
df_merged = pd.merge(
    new_subset,
    old_subset,
    on=['site_reference', 'year', 'month'],
    how='outer'
)

# 4. Consolidate nee_x and nee_y into a single, definitive 'nee' column
#    Fill missing values in nee_x with values from nee_y
df_merged['nee'] = df_merged['nee_x'].fillna(df_merged['nee_y'])
#    Drop the temporary columns
df_merged = df_merged.drop(columns=['nee_x', 'nee_y'])

# Create a date column from the merged data
df_merged['date'] = pd.to_datetime(df_merged[['year', 'month']].assign(day=1))


# 5. Define output path and get site list
comparison_plot_path = os.path.join("/explore/nobackup/people/spotter5/anna_v/v2/exploration", "ndvi_evi")
os.makedirs(comparison_plot_path, exist_ok=True)
print(f"Plots will be saved to: {comparison_plot_path}")

all_sites = df_merged['site_reference'].unique()
print(f"Found {len(all_sites)} unique sites to process.")


# --- 🛰️ Plotting Loop ---
for site in all_sites:
    if pd.isnull(site):
        continue

    # Filter for the site AND drop rows that do not have a valid observation in the *consolidated* NEE column
    site_df = df_merged[df_merged['site_reference'] == site].dropna(subset=['nee']).sort_values('date')

    # Skip site if it has no data left after filtering
    if site_df.empty:
        print(f"  -> Skipping site: {site} (No valid NEE observations)")
        continue
    
    print(f"  -> Processing site: {site}")

    # Create a 1x2 subplot figure
    fig, axes = plt.subplots(1, 2, figsize=(24, 9))

    # --- LEFT SUBPLOT: NEE vs NDVI ---
    ax1_ndvi = axes[0]
    ax2_ndvi = ax1_ndvi.twinx()

    line1 = ax1_ndvi.plot(site_df['date'], site_df['nee'], color='lightgray', label='NEE', linewidth=2.5, zorder=1)
    ax1_ndvi.set_ylabel('NEE ($gC/m^2/day$)', color='gray', fontsize=12)
    ax1_ndvi.tick_params(axis='y', labelcolor='gray')
    ax1_ndvi.set_xlabel('Date', fontsize=12)
    ax1_ndvi.set_title('NEE vs. NDVI', fontsize=14, fontweight='bold')
    ax1_ndvi.grid(True, linestyle='--', alpha=0.5)

    line2 = ax2_ndvi.plot(site_df['date'], site_df['NDVI_old'], color='cornflowerblue', linestyle='--', label='NDVI_old', zorder=2)
    line3 = ax2_ndvi.plot(site_df['date'], site_df['NDVI_new'], color='forestgreen', linestyle='-', label='NDVI_new', zorder=3)
    ax2_ndvi.set_ylabel('NDVI', color='darkgreen', fontsize=12)
    ax2_ndvi.tick_params(axis='y', labelcolor='darkgreen')

    lines_left = line1 + line2 + line3
    labels_left = [l.get_label() for l in lines_left]
    ax1_ndvi.legend(lines_left, labels_left, loc='upper left')

    # --- RIGHT SUBPLOT: NEE vs EVI ---
    ax1_evi = axes[1]
    ax2_evi = ax1_evi.twinx()

    line4 = ax1_evi.plot(site_df['date'], site_df['nee'], color='lightgray', label='NEE', linewidth=2.5, zorder=1)
    ax1_evi.set_ylabel('NEE ($gC/m^2/day$)', color='gray', fontsize=12)
    ax1_evi.tick_params(axis='y', labelcolor='gray')
    ax1_evi.set_xlabel('Date', fontsize=12)
    ax1_evi.set_title('NEE vs. EVI', fontsize=14, fontweight='bold')
    ax1_evi.grid(True, linestyle='--', alpha=0.5)

    line5 = ax2_evi.plot(site_df['date'], site_df['EVI_old'], color='lightcoral', linestyle='--', label='EVI_old', zorder=2)
    line6 = ax2_evi.plot(site_df['date'], site_df['EVI_new'], color='firebrick', linestyle='-', label='EVI_new', zorder=3)
    ax2_evi.set_ylabel('EVI', color='darkred', fontsize=12)
    ax2_evi.tick_params(axis='y', labelcolor='darkred')
    
    lines_right = line4 + line5 + line6
    labels_right = [l.get_label() for l in lines_right]
    ax1_evi.legend(lines_right, labels_right, loc='upper left')

    # --- Final Figure Formatting ---
    fig.suptitle(f"Site: {site}", fontsize=20, fontweight='bold')
    fig.tight_layout(rect=[0, 0, 1, 0.94])

    plot_filename = f"{site}_final_comparison.png"
    plt.savefig(os.path.join(comparison_plot_path, plot_filename), dpi=100)
    plt.close(fig)

print(f"\n--- ✅ Finished plotting. All site comparison plots are saved. ---")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import warnings

# Ignore routine warnings for cleaner output
warnings.filterwarnings('ignore')

# 1. Load your datasets
print("--- 📂 Loading datasets ---")
try:
    df_new = pd.read_csv("/explore/nobackup/people/spotter5/anna_v/v2/v2_model_training_alt_soil_lc_co2.csv")
    # df_new = pd.read_csv("/explore/nobackup/people/spotter5/anna_v/v2/v2_model_training_data_v2.csv")

    df_new = df_new[df_new['flux_method'] == 'EC']

    # df_new['NDVI'] = df_new['NDVI'] * 0.0001
    # df_new['EVI'] = df_new['EVI'] * 0.0001

    # --- UNITS CONVERSION FOR NEW DATA ---
    # Multiply by 0.1 to convert temperature to degrees Celsius
    df_new['tmean_C'] = df_new[['tmmn', 'tmmx']].mean(axis=1) #* 0.1
    # Convert srad from W/m^2 to MJ/m^2/day (1 W/m^2 = 0.0864 MJ/m^2/day)
    # df_new['srad'] = df_new['srad'] * 0.0864
    # df_new['srad'] = df_new['srad'] * 0.1
    # df_new['srad'] = df_new['srad'] * 0.1
    # df_new['site_reference'] = df_new['siteID']

    
    # df_old = pd.read_csv("/explore/nobackup/people/spotter5/anna_v/v2/v2_model_training_data_alt_soil_old.csv")
    df_old = pd.read_csv("/explore/nobackup/people/spotter5/anna_v/v2/v2_model_training_data_orig.csv")

    # Calculate mean temperature for the old dataset
    df_old['tmean_C'] = df_old[['tmax_C', 'tmin_C']].mean(axis=1)
    # Standardize column names
    df_old = df_old.rename(columns={'srad_MJ/m2': 'srad'})

    df_old['NDVI'] = df_old['NDVI'] * 0.0001
    df_old['EVI'] = df_old['EVI'] * 0.0001
    df_old['srad'] = df_old['srad'] * 0.1


except FileNotFoundError as e:
    print(f"Error: A data file was not found. Please check the path.\n{e}")
    exit()

# 2. Prepare dataframes for joining
# Include all variables of interest: nee, NDVI, EVI, tmean_C, srad
print("--- 🛠️ Preparing data for merging ---")
# Rename columns from the 'new' dataset to have the '_monthly' suffix
monthly_subset = df_new[['site_reference', 'year', 'month', 'nee', 'NDVI', 'EVI', 'tmean_C', 'srad']].rename(
    columns={'NDVI': 'NDVI_monthly', 'EVI': 'EVI_monthly', 'tmean_C': 'tmean_C_monthly', 'srad': 'srad_monthly'}
)
# Rename columns from the 'old' dataset to have the '_filled' suffix
filled_subset = df_old[['site_reference', 'year', 'month', 'nee', 'NDVI', 'EVI', 'tmean_C', 'srad']].rename(
    columns={'NDVI': 'NDVI_filled', 'EVI': 'EVI_filled', 'tmean_C': 'tmean_C_filled', 'srad': 'srad_filled'}
)

# 3. Join the dataframes. This creates '_x' (monthly) and '_y' (filled) suffixes
print("--- 🔗 Merging monthly and filled dataframes ---")
df_merged = pd.merge(
    monthly_subset,
    filled_subset,
    on=['site_reference', 'year', 'month'],
    how='outer'
)

# 4. Consolidate nee_x and nee_y into a single 'nee' column
df_merged['nee'] = df_merged['nee_x'].fillna(df_merged['nee_y'])
df_merged = df_merged.drop(columns=['nee_x', 'nee_y'])

# Create a date column from the merged data
df_merged['date'] = pd.to_datetime(df_merged[['year', 'month']].assign(day=1))


# 5. Define output path and get site list
comparison_plot_path = os.path.join("/explore/nobackup/people/spotter5/anna_v/v2/exploration", "variable_comparisons_v2")
os.makedirs(comparison_plot_path, exist_ok=True)
print(f"Plots will be saved to: {comparison_plot_path}")

all_sites = df_merged['site_reference'].unique()
print(f"Found {len(all_sites)} unique sites to process.")


# --- 🛰️ Plotting Loop ---
for site in all_sites:
    if pd.isnull(site):
        continue

    # Filter for the site AND drop rows without a valid consolidated NEE observation
    site_df = df_merged[df_merged['site_reference'] == site].dropna(subset=['nee']).sort_values('date')

    if site_df.empty:
        print(f" -> Skipping site: {site} (No valid NEE observations)")
        continue
    
    print(f" -> Processing site: {site}")

    # Create a 2x2 subplot figure
    fig, axes = plt.subplots(2, 2, figsize=(24, 18), sharex=True)
    fig.suptitle(f"Site: {site}", fontsize=20, fontweight='bold')

    # --- Plotting Function (to reduce code repetition) ---
    def create_comparison_plot(ax, site_data, var_name, var_unit, filled_color, monthly_color):
        ax_twin = ax.twinx()
        
        # Plot NEE on primary axis
        line_nee = ax.plot(site_data['date'], site_data['nee'], color='lightgray', label='NEE', linewidth=2.5, zorder=1)
        ax.set_ylabel('NEE ($gC/m^2/day$)', color='gray', fontsize=12)
        ax.tick_params(axis='y', labelcolor='gray')
        ax.grid(True, linestyle='--', alpha=0.5)

        # Plot variable on secondary axis
        line_filled = ax_twin.plot(site_data['date'], site_data[f'{var_name}_filled'], color=filled_color, linestyle='--', label=f'{var_name}_filled', zorder=2)
        line_monthly = ax_twin.plot(site_data['date'], site_data[f'{var_name}_monthly'], color=monthly_color, linestyle='-', label=f'{var_name}_monthly', zorder=3)
        ax_twin.set_ylabel(f'{var_name} ({var_unit})', color=monthly_color, fontsize=12)
        ax_twin.tick_params(axis='y', labelcolor=monthly_color)
        
        ax.set_title(f'NEE vs. {var_name}', fontsize=14, fontweight='bold')
        
        # --- Create a combined legend for both axes ---
        all_handles = line_nee + line_filled + line_monthly
        all_labels = [h.get_label() for h in all_handles]
        ax.legend(all_handles, all_labels, loc='upper left')

    # --- Generate the 4 plots ---
    # Top-Left: NDVI
    create_comparison_plot(axes[0, 0], site_df, 'NDVI', 'unitless', 'cornflowerblue', 'forestgreen')
    
    # Top-Right: EVI
    create_comparison_plot(axes[0, 1], site_df, 'EVI', 'unitless', 'lightcoral', 'firebrick')

    # Bottom-Left: Mean Temperature
    create_comparison_plot(axes[1, 0], site_df, 'tmean_C', '°C', 'skyblue', 'darkblue')
    
    # Bottom-Right: Solar Radiation
    create_comparison_plot(axes[1, 1], site_df, 'srad', 'MJ/m²/day', 'orange', 'saddlebrown')
    
    # --- Final Figure Formatting ---
    for ax in axes.flat:
        ax.set_xlabel('Date', fontsize=12)

    fig.tight_layout(rect=[0, 0, 1, 0.96]) # Adjust rect to make space for suptitle

    plot_filename = f"{site}_2x2_comparison.png"
    plt.savefig(os.path.join(comparison_plot_path, plot_filename), dpi=100)
    plt.close(fig)
    break

print(f"\n--- ✅ Finished plotting. All site comparison plots are saved. ---")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import warnings

# Ignore routine warnings for cleaner output
warnings.filterwarnings('ignore')

# 1. Load your datasets
print("--- 📂 Loading datasets ---")
try:
    df_new = pd.read_csv("/explore/nobackup/people/spotter5/anna_v/v2/v2_model_training_alt_soil_lc_co2.csv")
    df_old = pd.read_csv("/explore/nobackup/people/spotter5/anna_v/v2/v2_model_training_data_alt_soil_old.csv")
except FileNotFoundError as e:
    print(f"Error: A data file was not found. Please check the path.\n{e}")
    exit()

# 2. Select and rename the NEE columns from each dataframe
nee_new = df_new[['site_reference', 'year', 'month', 'nee']].rename(columns={'nee': 'nee_new'})
nee_old = df_old[['site_reference', 'year', 'month', 'nee']].rename(columns={'nee': 'nee_old'})

# 3. Join the dataframes on site and date
print("--- 🔗 Merging old and new NEE data ---")
nee_comparison_df = pd.merge(
    nee_new,
    nee_old,
    on=['site_reference', 'year', 'month'],
    how='outer'  # Outer join keeps all records from both files
)

# Create a date column for plotting
nee_comparison_df['date'] = pd.to_datetime(nee_comparison_df[['year', 'month']].assign(day=1))

# 4. Define output path and get site list
comparison_plot_path = os.path.join("/explore/nobackup/people/spotter5/anna_v/v2/exploration", "nee_value_comparison")
os.makedirs(comparison_plot_path, exist_ok=True)
print(f"Plots will be saved to: {comparison_plot_path}")

all_sites = nee_comparison_df['site_reference'].unique()
print(f"Found {len(all_sites)} unique sites to process.")


# --- 🛰️ Plotting Loop ---
for site in all_sites:
    if pd.isnull(site):
        continue

    # Filter for the current site
    site_df = nee_comparison_df[nee_comparison_df['site_reference'] == site].sort_values('date')

    # Skip site if it has no data at all
    if site_df.empty:
        print(f"  -> Skipping site: {site} (No data in merged file)")
        continue
    
    print(f"  -> Processing site: {site}")

    # Create a simple figure with one subplot
    fig, ax = plt.subplots(figsize=(18, 8))

    # Plot both NEE time series
    ax.plot(site_df['date'], site_df['nee_new'], label='NEE (New)', color='blue', marker='o', markersize=4, linestyle='-')
    ax.plot(site_df['date'], site_df['nee_old'], label='NEE (Old)', color='red', marker='x', markersize=5, linestyle='--')

    # Formatting
    ax.set_title(f"NEE Comparison for Site: {site}", fontsize=16, fontweight='bold')
    ax.set_xlabel("Date", fontsize=12)
    ax.set_ylabel("NEE ($gC/m^2/day$)", fontsize=12)
    ax.legend()
    ax.grid(True, linestyle='--', alpha=0.6)
    fig.tight_layout()

    # Save the figure to a PNG file
    plot_filename = f"{site}_nee_comparison.png"
    plt.savefig(os.path.join(comparison_plot_path, plot_filename), dpi=100)
    plt.close(fig)

print(f"\n--- ✅ Finished plotting. All site comparison plots are saved. ---")
```

```python
import pandas as pd

# Your initial setup
df_old = pd.read_csv("/explore/nobackup/people/spotter5/anna_v/v2/v2_model_training_data_alt_soil_old.csv")
df_old = df_old[['site_reference', 'nee']].dropna()

df_new = pd.read_csv("/explore/nobackup/people/spotter5/anna_v/v2/v2_model_training_data.csv")
df_new = df_new[df_new['flux_method'] == 'EC']
df_new = df_new[['site_reference', 'nee']].dropna()

## 1. Group and Count Observations

# Tally observations for each site in both DataFrames
# .size() is used here to count the number of rows for each group
old_counts = df_old.groupby('site_reference').size().rename('old_count')
new_counts = df_new.groupby('site_reference').size().rename('new_count')

## 2. Join Counts and Find the Difference

# Merge the two counts series into a new DataFrame.
# An 'outer' join includes all sites, even if they only appear in one DataFrame.
comparison_df = pd.merge(new_counts, old_counts, on='site_reference', how='outer')

# Replace NaN with 0 for sites not present in one of the lists and convert counts to integer
comparison_df = comparison_df.fillna(0).astype(int)

# Calculate the difference between the new and old observation counts
comparison_df['difference'] = comparison_df['new_count'] - comparison_df['old_count']

comparison_df = comparison_df.reset_index()
# --- Display Results ---
print("Comparison of observation counts per site:")
print(comparison_df)

# To see only the sites where the count has changed
print("\nSites with changes in observation count:")
print(comparison_df[comparison_df['difference'] != 0])

comparison_df.to_csv("/explore/nobackup/people/spotter5/anna_v/v2/count_compare.csv", index = False)
```

```python
import pandas as pd

# --- Inputs ---
CSV_IN  = "/explore/nobackup/people/spotter5/anna_v/v2/v2_model_training_final.csv"
RECLASS = "/explore/nobackup/people/spotter5/helene/raw/TEMLandcoverClassDictionary_v2.csv"

# --- Load and subset unique sites ---
df = pd.read_csv(CSV_IN)
df = df.drop_duplicates(subset="site_reference")
df = df[["site_reference", "latitude", "longitude", "land_cover_eco", "bawld_class", "land_cover"]].copy()

# --- Map land_cover_eco -> descriptive strings ---
landcover_eco_map = {
    40:  "Mosaic natural vegetation (>50%) / cropland (<50%)",
    60:  "Tree cover, broadleaved, deciduous",
    70:  "Tree cover, needleleaved, evergreen",
    80:  "Tree cover, needleleaved, deciduous",
    90:  "Tree cover, mixed leaf type",
    100: "Mosaic tree & shrub (>50%) / herbaceous (<50%)",
    110: "Mosaic herbaceous (>50%) / tree & shrub (<50%)",
    120: "Shrubland",
    121: "Shrubland evergreen",
    122: "Shrubland deciduous",
    130: "Grassland",
    140: "Lichens and mosses",
    150: "Sparse vegetation (<15%)",
    151: "Sparse tree (<15%)",
    152: "Sparse shrub (<15%)",
    153: "Sparse herbaceous (<15%)",
    160: "Tree cover, flooded, fresh or brackish water",
    170: "Tree cover, flooded, saline water",
    180: "Shrub/herbaceous cover, flooded (fresh/saline/brackish)",
    200: "Bare areas",
    210: "Water bodies",
}
df["land_cover_eco_code"] = pd.to_numeric(df["land_cover_eco"], errors="coerce").astype("Int64")
df["land_cover_eco"] = df["land_cover_eco_code"].map(landcover_eco_map)

# --- Map land_cover (TEM v2) using CSV: Reclass -> New Name ---
# Robust read in case of tabs/multiple spaces
class_df = pd.read_csv(RECLASS, sep=None, engine="python")
# Normalize headers
class_df.columns = class_df.columns.str.strip()

# Expecting these exact columns from your sample:
# ["Value", "Class Name", "Reclass", "New Name"]
if not {"Reclass", "New Name"}.issubset(class_df.columns):
    raise ValueError(f"CSV must contain 'Reclass' and 'New Name' columns. Found: {list(class_df.columns)}")

# Build mapping dict (Reclass code -> New Name)
class_df["Reclass"] = pd.to_numeric(class_df["Reclass"], errors="coerce").astype("Int64")
reclass_to_name = (
    class_df.dropna(subset=["Reclass", "New Name"])
            .drop_duplicates(subset=["Reclass"])
            .set_index("Reclass")["New Name"]
            .astype(str)
            .to_dict()
)

# Apply mapping
df["land_cover_code_v2"] = pd.to_numeric(df["land_cover"], errors="coerce").astype("Int64")
df["land_cover"] = df["land_cover_code_v2"].map(reclass_to_name)

# --- Optional: report unmapped codes ---
missing_tem = df.loc[df["land_cover"].isna(), "land_cover_code_v2"].dropna().unique()
missing_eco = df.loc[df["land_cover_eco"].isna(), "land_cover_eco_code"].dropna().unique()
if len(missing_tem):
    print("⚠️ Unmapped TEM v2 land_cover codes:", sorted(map(int, missing_tem)))
if len(missing_eco):
    print("⚠️ Unmapped land_cover_eco codes:", sorted(map(int, missing_eco)))

df.to_csv("/explore/nobackup/people/spotter5/anna_v/v2/all_landcover_classes.csv", index = False)
```

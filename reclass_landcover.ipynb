```python
import rioxarray as rxr
import pandas as pd
import numpy as np
import rasterio

# Paths
landcover_raster_path = "/explore/nobackup/people/spotter5/helene/raw/TEM_Landcover_V4.tif"
reclass_csv ="/explore/nobackup/people/spotter5/helene/raw/TEMLandcoverClassDictionary.csv"
output_raster_path = "/explore/nobackup/people/spotter5/anna_v/v2/TEM_Landcover_Reclassified.tif"
nodata_value = -9999

# Load landcover raster
landcover_raster = rxr.open_rasterio(landcover_raster_path, masked=True).squeeze()
landcover_values = landcover_raster.values
transform = landcover_raster.rio.transform()
crs = landcover_raster.rio.crs
shape = landcover_values.shape

# Load and prepare reclassification mapping
class_df = pd.read_csv(reclass_csv)
class_df.columns = class_df.columns.str.strip()
class_df['Class Name'] = class_df['Class Name'].str.strip()
reclass_map = class_df.set_index('Value')['Reclass'].dropna().astype(int).to_dict()

# Reclassify
reclassified = np.full_like(landcover_values, np.nan, dtype="float32")
for orig, new in reclass_map.items():
    reclassified[landcover_values == orig] = new

# Save reclassified raster
reclassified_to_save = np.where(np.isnan(reclassified), nodata_value, reclassified).astype("int16")

with rasterio.open(
    output_raster_path,
    "w",
    driver="GTiff",
    height=shape[0],
    width=shape[1],
    count=1,
    dtype="int16",
    crs=crs,
    transform=transform,
    nodata=nodata_value
) as dst:
    dst.write(reclassified_to_save, 1)

print(f"✅ Reclassified landcover raster saved to: {output_raster_path}")
```

```python
import geopandas as gpd
import rioxarray as rxr
import pandas as pd
import xarray as xr

# Define the paths for your input files and the desired output file
shapefile_path = "/explore/nobackup/people/spotter5/anna_v/v2/v2_model_sites.shp"
tif_path = "/explore/nobackup/people/spotter5/helene/raw/TEM_Landcover_V4.tif"
reclass_csv_path = "/explore/nobackup/people/spotter5/helene/raw/TEMLandcoverClassDictionary.csv"
output_csv_path = "/explore/nobackup/people/spotter5/anna_v/v2/extracted_landcover_values_original.csv"

# 1. Load the shapefile, raster file, and reclassification table
print("Loading data...")
sites_gdf = gpd.read_file(shapefile_path)
landcover_raster = rxr.open_rasterio(tif_path).squeeze()
reclass_df = pd.read_csv(reclass_csv_path)
reclass_df = reclass_df.rename(columns = {'Class Name ': 'Class_Name'})
reclass_df = reclass_df[['Value', 'Class_Name']]

# 2. Ensure Coordinate Reference Systems (CRS) match
print("Aligning coordinate systems...")
sites_reprojected = sites_gdf.to_crs(landcover_raster.rio.crs)

# 3. Extract raster values at point locations
print("Extracting raster values...")

# Create DataArrays for the x and y coordinates of the points
x_coords = xr.DataArray(sites_reprojected.geometry.x, dims="point")
y_coords = xr.DataArray(sites_reprojected.geometry.y, dims="point")

# Use .sel() to select the raster values at the nearest grid cell for each point
extracted_values = landcover_raster.sel(
    x=x_coords, y=y_coords, method="nearest"
)

# 4. Add extracted values to the GeoDataFrame
sites_reprojected['land_cover_value'] = extracted_values.values

# 5. Merge the class names from the reclassification DataFrame
print("Merging land cover class names...")
# We perform a 'left' merge to keep all the sites from the shapefile
# and add the corresponding class name based on the extracted raster value.
sites_with_landcover = sites_reprojected.merge(
    reclass_df,
    left_on='land_cover_value',
    right_on='Value',
    how='left'
)

# 6. Save the results to a CSV file
print(f"Saving results to {output_csv_path}...")
# Drop the redundant 'Value' column and the 'geometry' column before saving
final_df = pd.DataFrame(sites_with_landcover.drop(columns=['geometry', 'Value']))
final_df.to_csv(output_csv_path, index=False)

print("\nProcessing complete.")
print("First 5 rows of the output:")
print(final_df.head())
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------- BIGGER TEXT (global defaults) -----------------
plt.rcParams.update({
    "font.size": 14,          # base font size
    "axes.titlesize": 22,     # axes title
    "axes.labelsize": 16,     # x/y labels
    "xtick.labelsize": 13,    # tick labels
    "ytick.labelsize": 13,
    "legend.fontsize": 14,
    "figure.titlesize": 24
})
sns.set_context("talk", font_scale=1.1)  # extra bump for seaborn

# ----------------- INPUTS -----------------
# Original extraction + TEM class dictionary + flux info (as in your code)
extracted_values_csv = "/explore/nobackup/people/spotter5/anna_v/v2/extracted_landcover_values_original.csv"
reclass_csv_path     = "/explore/nobackup/people/spotter5/helene/raw/TEMLandcoverClassDictionary.csv"
flux_data_path       = "/explore/nobackup/people/spotter5/anna_v/v2/v2_model_training_final.csv"

# New classification source + mapping (your remap; here it is the *Original* system)
v3_plot_csv          = "/explore/nobackup/people/spotter5/anna_v/v2/v2_model_training_data_v3.csv"

# ----------------- OUTPUTS -----------------
out_dir = "/explore/nobackup/people/spotter5/anna_v/v2"
os.makedirs(out_dir, exist_ok=True)

output_ec_sites_csv   = os.path.join(out_dir, "extracted_landcover_ec_sites.csv")
plot_ec_pct_png       = os.path.join(out_dir, "landcover_distribution_ec_sites_percent.png")

# Crosswalk outputs (Original -> TEM (New))
crosswalk_counts_csv  = os.path.join(out_dir, "crosswalk_ORIGINAL_landcover_plot_to_TEMNEW_counts.csv")
crosswalk_rowpct_csv  = os.path.join(out_dir, "crosswalk_ORIGINAL_landcover_plot_to_TEMNEW_rowpct.csv")
crosswalk_heatmap_png = os.path.join(out_dir, "crosswalk_ORIGINAL_to_TEMNEW_rowpct_heatmap_REDS.png")

# ----------------- LOAD ORIGINAL + TEM (NEW) -----------------
print("Loading pre-extracted data...")
sites_df   = pd.read_csv(extracted_values_csv)
reclass_df = pd.read_csv(reclass_csv_path)

# Clean reclass table (TEM dictionary): {Value -> Class_Name}
reclass_df = reclass_df.rename(columns={'Class Name ': 'Class_Name'})
reclass_df = reclass_df[['Value', 'Class_Name']]

# Map original site raster values to TEM class names
sites_df = sites_df.rename(columns={'land_cover': 'land_cover_value'})
if 'Class_Name' in sites_df.columns:
    sites_df = sites_df.drop(columns=['Class_Name'])

sites_with_landcover = sites_df.merge(
    reclass_df, left_on='land_cover_value', right_on='Value', how='left'
)  # Class_Name = TEM (this is the NEW classification)

# ----------------- LOAD FLUX + FILTER EC -----------------
print("Loading flux data and filtering for EC sites...")
flux_df = pd.read_csv(flux_data_path).drop_duplicates(subset=['site_reference'])
flux_df = flux_df[['site_reference', 'flux_method']]

# Normalize site id column on the TEM side
sites_with_landcover = sites_with_landcover.rename(columns={'site_refer': 'site_reference'})
merged_df = sites_with_landcover.merge(flux_df, on='site_reference', how='inner')
ec_sites_df = merged_df[merged_df['flux_method'] == 'EC'].copy()
final_df    = ec_sites_df.drop(columns=['Value'], errors='ignore')
# final_df.to_csv(output_ec_sites_csv, index=False)  # optional

# ----------------- PLOT TEM (NEW) CLASS % FOR EC SITES -----------------
print(f"Generating TEM (NEW) class % bar plot → {plot_ec_pct_png}")
tem_percent = ec_sites_df['Class_Name'].value_counts(normalize=True) * 100
tem_percent_df = tem_percent.rename_axis('TEM_Class_Name').reset_index(name='Percentage')

fig, ax = plt.subplots(figsize=(14, 12))
sns.set_style("whitegrid")
sns.barplot(x='Percentage', y='TEM_Class_Name', data=tem_percent_df, palette='viridis', ax=ax)
ax.set_title('TEM (NEW) Land Cover Class Distribution for EC Sites (%)', fontsize=22)
ax.set_xlabel('Percentage of Pixels (%)', fontsize=16)
ax.set_ylabel('TEM Class Name', fontsize=16)
ax.tick_params(axis='both', labelsize=13)
plt.tight_layout()
plt.savefig(plot_ec_pct_png, dpi=300)
plt.close(fig)

# ----------------- LOAD ORIGINAL SYSTEM + REMAP land_cover_plot -> names -----------------
print("Loading ORIGINAL (plot) classification and applying remap...")
plot_df = pd.read_csv(v3_plot_csv)

landcover_map = {
    40:  "Mosaic natural vegetation (>50%) / cropland (<50%)",
    60:  "Tree cover, broadleaved, deciduous",
    70:  "Tree cover, needleleaved, evergreen",
    80:  "Tree cover, needleleaved, deciduous",
    90:  "Tree cover, mixed leaf type",
    100: "Mosaic tree & shrub (>50%) / herbaceous (<50%)",
    110: "Mosaic herbaceous (>50%) / tree & shrub (<50%)",
    120: "Shrubland",
    121: "Shrubland evergreen",
    122: "Shrubland deciduous",
    130: "Grassland",
    140: "Lichens and mosses",
    150: "Sparse vegetation (<15%)",
    151: "Sparse tree (<15%)",
    152: "Sparse shrub (<15%)",
    153: "Sparse herbaceous (<15%)",
    160: "Tree cover, flooded, fresh or brackish water",
    170: "Tree cover, flooded, saline water",
    180: "Shrub/herbaceous cover, flooded (fresh/saline/brackish)",
    200: "Bare areas",
    210: "Water bodies",
}

if 'land_cover_plot' not in plot_df.columns:
    raise KeyError("Column 'land_cover_plot' not found in the new classification CSV.")

plot_df['land_cover_plot']      = pd.to_numeric(plot_df['land_cover_plot'], errors='coerce')
plot_df['land_cover_plot_name'] = plot_df['land_cover_plot'].map(landcover_map).fillna("Unknown")

# Normalize site id column on the ORIGINAL side
if 'site_reference' not in plot_df.columns and 'site_refer' in plot_df.columns:
    plot_df = plot_df.rename(columns={'site_refer': 'site_reference'})

# ----------------- FIND JOIN KEYS (pixel-aware if possible) -----------------
preferred_join_keys = [
    ['site_reference', 'pixel_id'],
    ['site_reference', 'plot_id'],
    ['site_reference', 'row', 'col'],
    ['site_reference', 'x', 'y'],
    ['site_reference']  # fallback
]

available_join = None
for keys in preferred_join_keys:
    if all(k in ec_sites_df.columns for k in keys) and all(k in plot_df.columns for k in keys):
        available_join = keys
        break

if available_join is None:
    raise RuntimeError("No suitable join keys found between EC site table and v3 plot table.")

if available_join == ['site_reference']:
    print("WARNING: Joining only on 'site_reference'. "
          "Many-to-many at the site level may mix pixels across classes.")

print(f"Joining with keys: {available_join}")

# Keep only needed columns for the join + labels
left_keep  = available_join + ['Class_Name']            # TEM (NEW)
right_keep = available_join + ['land_cover_plot_name']  # ORIGINAL

left_df  = ec_sites_df[left_keep].copy()
right_df = plot_df[right_keep].copy()

# INNER join keeps overlap only
crosswalk_df = left_df.merge(right_df, on=available_join, how='inner', validate='m:m')

# ----------------- BUILD TRANSITION MATRIX -----------------
# ORIGINAL = land_cover_plot_name (rows)
# NEW (TEM) = Class_Name (cols)
counts = pd.crosstab(
    crosswalk_df['land_cover_plot_name'],   # ORIGINAL system
    crosswalk_df['Class_Name'],             # TEM (NEW)
    dropna=False
)

# Row-normalized (%) -> "what % of ORIGINAL class goes to each TEM class"
row_pct = counts.div(counts.sum(axis=1).replace(0, np.nan), axis=0) * 100.0

# Save
counts.to_csv(crosswalk_counts_csv)
row_pct.to_csv(crosswalk_rowpct_csv)
print(f"Saved crosswalk counts → {crosswalk_counts_csv}")
print(f"Saved crosswalk row %  → {crosswalk_rowpct_csv}")

# ----------------- HEATMAP (ROW %) — REDS (larger text) -----------------
fig, ax = plt.subplots(
    figsize=(
        max(12, 0.45 * row_pct.shape[1] + 6),
        max(10, 0.45 * row_pct.shape[0] + 6)
    )
)
sns.set(font_scale=1.0)  # keep consistent; rcParams already bumped
hm = sns.heatmap(
    row_pct.fillna(0.0),
    cmap="Reds",
    annot=False,
    cbar_kws={'label': 'Row % (Original → TEM)'},
    ax=ax
)
ax.set_title("Site Level-TEM%", fontsize=22, pad=12)
ax.set_xlabel("TEM", fontsize=16)
ax.set_ylabel("Site Level", fontsize=16)
ax.tick_params(axis='both', labelsize=13)

# make colorbar text bigger too
cbar = hm.collections[0].colorbar
cbar.ax.tick_params(labelsize=13)
cbar.set_label('% of Sites', fontsize=16)

plt.tight_layout()
plt.savefig(crosswalk_heatmap_png, dpi=300)
plt.show()
plt.close(fig)
print(f"Saved crosswalk heatmap → {crosswalk_heatmap_png}")

# ----------------- EXAMPLE LOOKUP -----------------
# Example: percent from Original 'Shrubland' to TEM 'Larch Forest'
orig = "Shrubland"
new  = "Larch Forest"  # must match a TEM Class_Name string
if orig in row_pct.index and new in row_pct.columns:
    val = float(row_pct.loc[orig, new])
    print(f"\nPercent of pixels from '{orig}' (Original) → '{new}' (TEM NEW): {val:.2f}%")
else:
    print("\nExample lookup note: Either the Original class 'Shrubland' "
          "or the TEM class 'Larch Forest' was not found. "
          "Check the exact TEM class name (as listed in the TEM dictionary).")

print("\nDone.")
```
